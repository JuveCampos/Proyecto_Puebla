# Mapa de resultados electorales de Puebla

En este notebook, explicaremos el proceso necesario para elaborar un mapa basico en Python. (Python 3). 

```python
# Import packages
import geopandas as gpd
import matplotlib.pyplot as plt

# Abrimos el archivo .geojson
Puebla = gpd.read_file("DatosGeograficos/Puebla.geojson")

# Imprimimos los 5 primeros renglones
print(Puebla.head(5))
```

```python
# Recodificando la variable

# 1. Generamos un diccionario con los nombres completos de cada candidato
candidatos = {"MEAH": "Martha Erika Alonso Hidalgo (PAN, PRD, MC, CPP, PSI)",
             "LMGBH": "Luis Miguel Geronimo Barbosa Huerta(PT, MOR, PES)",
             "JEDG" : "Jose Enrique Doger Guerrero (PRI)",
             "MCC"  : "Miguel Chain Carrillo (P"}

colores = {"MEAH": 'blue', 
           "LMGBH": 'red', 
           "JEDG": 'green', 
           "MCC": 'yellow'}

# 2. Generamos una nueva variable con la funcion map y el diccionario que creamos antes
#      Lo que hace esta funcion es que a cada valor de "GANADOR" le asigna el valor que tiene en el diccionario
#      y se lo da a otra variable.
Puebla['Ganador'] = Puebla.GANADOR.map(candidatos)

Puebla['Color'] = Puebla.GANADOR.map(colores)

# 3. Imprimimos el contenido
print(Puebla.head())
```

```python
# HACEMOS EL MAPA

# Graficamos los poligonos
Puebla.plot(column = 'Ganador', legend = True)

# Agregamos etiquetas a los ejes y titulo
plt.title("Ganador de las elecciones a gobernador por municipio")
plt.xlabel('longitud')
plt.ylabel('latitud')

# Add grid lines and show the plot
plt.grid()
plt.show()
```

```python
f, ax = plt.subplots(1, figsize=(11, 11))
ax = Puebla.plot(column = 'Ganador', legend = True, ax = ax)
plt.title("Ganador de las elecciones a gobernador por municipio")
ax.set_axis_off()
plt.show()
```
